```python
# PREGUNTA 2
minCO2Mundo=carbon_copy[carbon_copy.metric_tonnes_of_CO2==min(carbon_copy.metric_tonnes_of_CO2)]
minCO2Mundo
```

```python
# PREGUNTA 3
minCO2Region=carbon_copy.loc[carbon_copy.groupby('region')['metric_tonnes_of_CO2'].idxmin()]
minCO2Region
```

```python
# PREGUNTA 4
carbon2=carbon_copy.copy()

def potencia(x):
  result=x**0.5
  return result

carbon2['sqrt_metric_tonnes_of_CO2']=carbon2.metric_tonnes_of_CO2.apply(potencia)
carbon2
```

```python
# PREGUNTA 5
carbon3=carbon_copy.copy()
promCO2=carbon3['metric_tonnes_of_CO2'].mean()
carbon3['original-average']=carbon3['metric_tonnes_of_CO2']-promCO2
carbon3
```
